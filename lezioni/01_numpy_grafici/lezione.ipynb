## 2. Array NumPy vs liste Python

Una **lista Python** normale e un **array NumPy** sembrano simili, ma si comportano in modo molto diverso quando applichiamo operazioni matematiche.

````python
# Lista Python normale
lista = [1, 2, 3, 4, 5]

# Array NumPy
arr = np.array([1, 2, 3, 4, 5])

# Proviamo ad elevare al quadrato
print('lista * 2  =', lista * 2)   # duplica la lista!
print('arr ** 2   =', arr ** 2)    # eleva ogni elemento al quadrato ✓
print()
print('Tipo lista:', type(lista))
print('Tipo arr:  ', type(arr))
````

## 3. `np.linspace` — campionare un intervallo

`np.linspace(start, stop, num)` genera `num` valori **ugualmente spaziati** tra `start` e `stop`.

```python
np.linspace(0, 1, 5)   →  [0.0,  0.25,  0.5,  0.75,  1.0]
```

È lo strumento perfetto per costruire il vettore delle **x** di un grafico.

````python
# 6 punti tra 0 e 5
x_piccolo = np.linspace(0, 5, 6)
print('6 punti  :', x_piccolo)
print('Lunghezza:', len(x_piccolo))
print('Passo tra i valori:', x_piccolo[1] - x_piccolo[0])
````

## 4. Il primo grafico

Per disegnare una funzione:
1. Crea il vettore `x` con `linspace`
2. Calcola `y = f(x)` — NumPy opera su tutti gli elementi in una riga
3. Usa `plt.plot(x, y)` per disegnare
4. Aggiungi titolo, legenda, griglia
5. Mostra con `plt.show()`

````python
x = np.linspace(0, 5, 30)

plt.figure('tre funzioni', figsize=[5, 3.5])

plt.plot(x, x,     label='lineare',    linewidth=2)
plt.plot(x, x**2,  label='quadratica', linewidth=2)
plt.plot(x, x**3,  label='cubica',     linewidth=2)

plt.title('Grafico di tre funzioni con NumPy')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, color='grey')
plt.tight_layout()
plt.show()
````

---
## 5. Effetto del numero di punti

Il terzo argomento di `linspace` controlla quanti punti usi per disegnare la curva.  
Con pochi punti il grafico appare spezzato; con molti punti diventa liscio.

````python
fig, assi = plt.subplots(1, 3, figsize=(12, 3.5))

for ax, n in zip(assi, [5, 30, 300]):
    x = np.linspace(0, 5, n)
    ax.plot(x, x**2, marker='o', markersize=4)
    ax.set_title(f'linspace(0, 5, {n})')
    ax.set_xlabel('x')
    ax.set_ylabel('x²')
    ax.grid(True, alpha=0.4)

plt.suptitle('Effetto del numero di punti', fontweight='bold')
plt.tight_layout()
plt.show()
````

## 6. Personalizzare il grafico

Matplotlib offre molti parametri per personalizzare ogni curva:

| Parametro | Valori tipici | Effetto |
|-----------|---------------|---------|
| `color` | `'red'`, `'#2e7dc9'`, `'C2'` | Colore della curva |
| `linewidth` | `1`, `2`, `3` | Spessore della linea |
| `linestyle` | `'-'`, `'--'`, `':'`, `'-.'` | Stile della linea |
| `label` | stringa | Etichetta nella legenda |
| `alpha` | `0.0` – `1.0` | Trasparenza |

````python
x = np.linspace(0, 5, 200)

plt.figure(figsize=(6, 4))

plt.plot(x, x,     color='#2e7dc9', linewidth=2,  linestyle='-',  label='lineare')
plt.plot(x, x**2,  color='red',     linewidth=2,  linestyle='--', label='quadratica')
plt.plot(x, x**3,  color='#1f8a5e', linewidth=2,  linestyle=':',  label='cubica')

plt.title('Personalizzazione del grafico')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## 7. Funzioni matematiche di NumPy

NumPy include tutte le funzioni matematiche fondamentali, e le applica su tutto il vettore in un colpo solo:

```python
np.sqrt(x)   # radice quadrata
np.sin(x)    # seno (argomento in radianti)
np.cos(x)    # coseno
np.exp(x)    # e^x
np.log(x)    # logaritmo naturale
np.tan(x)    # tangente  ← attenzione alle discontinuità!
```

> 🔑 **Costante π:** usa `np.pi` — è già disponibile in NumPy.

````python
# Due vettori con intervalli diversi
x_pos  = np.linspace(0.01, 5, 300)          # x > 0 per sqrt e log
x_trig = np.linspace(0, 2 * np.pi, 300)     # un periodo completo per sin/cos

fig, assi = plt.subplots(1, 3, figsize=(13, 4))

# Radice quadrata
assi[0].plot(x_pos, np.sqrt(x_pos), color='#2e7dc9', linewidth=2)
assi[0].set_title('√x')
assi[0].grid(True, alpha=0.3)

# Seno
assi[1].plot(x_trig, np.sin(x_trig), color='#1f8a5e', linewidth=2)
assi[1].set_title('sin(x)  su [0, 2π]')
assi[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
assi[1].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
assi[1].grid(True, alpha=0.3)

# Logaritmo
assi[2].plot(x_pos, np.log(x_pos), color='#e07b2a', linewidth=2)
assi[2].set_title('ln(x)')
assi[2].grid(True, alpha=0.3)

plt.suptitle('Funzioni matematiche in NumPy', fontweight='bold')
plt.tight_layout()
plt.show()
````

## 8. ⚠️ Caso critico: la tangente

La tangente ha **asintoti verticali** in $x = \frac{\pi}{2} + k\pi$.  
Se usiamo un vettore che attraversa questi punti, Matplotlib collegherà i valori con una linea verticale che non corrisponde alla matematica reale.

Esegui la cella qui sotto e osserva cosa succede.

````python
x = np.linspace(0, 2 * np.pi, 300)
y = np.tan(x)

plt.figure(figsize=(7, 4))
plt.plot(x, y, color='red', linewidth=1.5)
plt.ylim(-10, 10)   # limita l'asse y per rendere leggibile il grafico
plt.axhline(0, color='black', linewidth=0.8)
plt.title('tan(x) — nota le righe verticali spurie!')
plt.xlabel('x')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print('Valore massimo calcolato:', np.max(np.abs(y)))
````

### Come correggere il problema

Una soluzione è **mascherare** i valori troppo grandi in valore assoluto, sostituendoli con `NaN` (Not a Number).  
Matplotlib non disegna i segmenti che collegano a un `NaN`, quindi le false verticali scompaiono.

````python
x = np.linspace(0, 2 * np.pi, 3000)  # più punti → migliore approssimazione
y = np.tan(x)

# Maschera: mette NaN dove |tan(x)| > 10
y[np.abs(y) > 10] = np.nan

plt.figure(figsize=(7, 4))
plt.plot(x, y, color='#2e7dc9', linewidth=2)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.8)
plt.title('tan(x) — versione corretta con maschera NaN')
plt.xlabel('x')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
